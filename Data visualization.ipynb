```python
#using plot() function to plot random dates on line graph

import pandas as pd
import numpy as np
df=pd.DataFrame(np.random.randn(10,4),index=pd.date_range('1/1/2000',periods=10),columns=list('ABCD'))

print(df)

df.plot()
```

```python
#using plot() function to plot random dates

import pandas as pd
import numpy as np
df=pd.DataFrame(np.random.randn(10,2),index=pd.date_range('1/1/2000',periods=10),columns=list(['A','B']))

print(df)

df.plot()
```

```python
#using plot.bar() function to plot data on bar graph

import pandas as pd
import numpy as np
df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])

print(df)

df.plot.bar()
```

```python
#using plot.bar() function to plot data on bar graph with 'stacked' attribute

import pandas as pd

df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])

print(df)
df.plot.bar(stacked=True)
```

```python
#using plot.barh() function to plot data on  horizontal bar graph with 'stacked' attribute

import pandas as pd

df=pd.DataFrame(np.random.rand(8,4),columns=['a','b','c','d'])

print(df)

df.plot.barh(stacked=True)

df.plot.barh(stacked=False)
```

```python
#using plot.barh() function to plot data on horizontal bar graph with 'stacked' attribute

import pandas as pd
import numpy as np

data={'animal':['cat','dog','goat','snake','cat','dog','horse','cow','goat','sheep'],
     'age':[1,2,3,1,2,3,1,2,1,1]}
df=pd.DataFrame(data,index=('a','b','c','d','e','f','g','h','i','j'))

print(df)

df.plot.bar(stacked=False)
```

```python
#plot.hist() function for histogram graph

import pandas as pd
import numpy as np

df=pd.DataFrame({'a':np.random.randn(1000)+1,'b':np.random.randn(1000),'c':np.random.randn(1000)-1},columns=['a','b','c'])

print(df)

df.plot.hist(bins=20)
```

```python
#using plot.box() function to plot box graph

import pandas as pd

df=pd.DataFrame(np.random.rand(8,4),columns=['a','b','c','d'])

print(df)

df.plot.box()
```

```python
#using plot.area() function

import pandas as pd

df=pd.DataFrame(np.random.rand(8,4),columns=['a','b','c','d'])

print(df)

df.plot.area()
```

```python
#using plot.scatter() function
import pandas as pd
import numpy as np

df=pd.DataFrame(np.random.rand(10,5),columns=['a','b','c','d','e'])

print(df)


df.plot.scatter(x='a',y='d')
```

```python
#using plot.pie() function to plot pie graph
import pandas as pd
import numpy as np

df=pd.DataFrame(3 * np.random.rand(4),index=['a','b','c','d'],columns=['x']) 
print(df)

df.plot.pie(subplots=True)
```
